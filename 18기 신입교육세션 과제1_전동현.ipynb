# 관심 있는 주제 : 경제

## 데이터 1
* 데이터명 : 소비자물가지수_15193802
* 데이터 특징
  * 출처 : 한국은행 경제통계시스템(https://ecos.bok.or.kr/#/SearchStat)
  * 분야 : 경제
  * 내용
    * 본 데이터는 2022년 1월부터 2025년 12월까지의 월별 소비자물가지수를 담고 있습니다. 첫 번째 행에는 월별 총지수를 나타내고 있으며 2020년을 100이라고 가정했을 때 해당 월의 소비자물가지수는 어느정도인지 수치상으로 나타냅니다.
    * 소비자물가지수는 가계가 구입하는 대표적인 상품과 서비스의 가격 변동을 종합적으로 나타내는 지표로, 물가 수준과 인플레이션 추이를 파악하는 데 활용됩니다.
    * 현재 날짜 기준 전월인 12월까지의 데이터를 담고 있으므로 비교적 최근 데이터라고 할 수 있으며 최근 물가 변동과 경제 환경 변화를 분석하는 데 유용합니다.
    * 따라서 최근들어 체감이 되는 인플레이션이 정말 수치상으로 나타나는 물가상승인지를 확인할 수 있으며 2020년 대비 어느 정도의 물가 상승이 있었는지를 확인할 수 있습니다.

## 데이터 2
* 데이터명 : 국세청_근로소득 백분위(천분위) 자료_20241231
* 데이터 특징
  * 출처 : 공공데이터 포털(https://www.data.go.kr/data/15082063/fileData.do)
  * 분야 : 금융
  * 내용
    * 본 데이터는 국세청에서 제공한 근로소득 백분위 자료로, 근로자의 소득 수준별 인원 수와 총급여, 근로소득금액, 소득공제액 등 근로 소득 관련 핵심 지표를 포함하고 있습니다. 이를 통해 소득 상위 계층으로 갈수록 급여 수준과 소득 구조가 어떻게 변화하는지 확인할 수 있습니다. 특히 총급여와 근로소득금액의 차이를 비교하여 비과제 항목 혹은 공제 항목의 영향을 간접적으로 파악할 수 있습니다.
    * 백분위 간 소득 격차를 비교함으로써 소득 불평등 및 양극화 정도를 파악할 수 있습니다. 또한 <데이터 1>의 소비자물가지수와 결합하면 실질소득 변화와 인플레이션의 영향을 분석하는 데 활용할 수 있습니다.
